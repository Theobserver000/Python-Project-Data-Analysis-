```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline 
import seaborn as sns 
```

```python
df=pd.read_csv('Diwali Sales Data.csv',encoding='unicode_escape')
```

```python
df.shape # this will result in output of how many rows and colomns 
```

```python
df.head(10)
```

```python
df.info() # this will givr us mini discription of tables
# from this we come to know that status and unmaned colomns name should be deleted
```

```python
df.drop(['Status','unnamed1'],axis=1,inplace=True)
```

```python
df.info()
```

```python
df.isnull() # isne surat ki kuch or last ki kuch values dikha din ki dekho false hain matln null value nahi hain par hame abhi bhi pura deta ki  jankari nahi mili hain
```

```python
df.isnull().sum() # isne pura calculate krke bata diya kii amount me totoal 12 null values hain 
```

```python
df.dropna(inplace=True)# yeh jitni bhi null values vali rows hongi unhe hata dega 
```

```python
df.shape
```

```python
df
```

```python
df.info()
```

```python
# We are going to change the data types of some colomns as like Amount ,

```

```python
df['Amount'].dtype # we can see that their data tyoe of float64
```

```python
df['Amount']=df['Amount'].astype('int') # we are changing here the data type of Amount
```

```python
df['Amount'].dtype # data type has been changed 
```

```python
df.columns # iske through hame pata chalta hain ki df ke andar kon kon se columns hain 
```

```python
df.rename(columns={'Marital_Status': 'Shadi_ka_status'}) # this could change the name of columns but this will not save as this because we are not using 
          #inplace=True so mean is that action has been performed but after action that action has not saved in the table so that action will not cause effect to original table 
          
```

```python
df.columns # yeh dekhiye hamne inplace nhi lagay toh koi actual change hua hi nhi
```

```python
df.describe() 
# here are some basic calculation over the data 




```

```python
ax=sns.countplot(x='Gender',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
# now from here we are using seaborn visualizing the data here we are giving x axis as genders and data for graph 
# is come from data=df 
ax
```

```python
sales_gen=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Gender', y='Amount',data=sales_gen)
```

From the above graphs we can clearly see that most of the buyers are females and even the purchasing power of females are greater than men 

```python
df.columns 
```

```python
ax=sns.countplot(x='Age Group',data=df,hue='Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Age Group' , y='Amount', data=sales_age)
```

From above graphs we can see that most of the buyers are of age group between 26-35 yrs female

```python
sales_state=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x='State',y='Orders',data=sales_state)

# this will only show top ten states as head(10)
#15,5 is the size of graphs set to not overlap values between each other 
```

```python
sales_state_amount=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x='State',y='Amount',data=sales_state_amount)
```

from the above graphs we can see that most of the orders from Uttar Pradesh 

```python
ax =sns.countplot(data=df,x='Marital_Status')
sns.set(rc={'figure.figsize':(3,4)})

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state_gender=df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(x='Marital_Status',y='Amount',data=sales_state_gender,hue='Gender')
```

From the above graphs we can see that most of the married women are shoping in big numbers and has high purchasing power 

```python
sns.set(rc={'figure.figsize':(20,5)})
ax =sns.countplot(data=df,x='Occupation')
for bars in ax.containers:
    ax.bar_label(bars)

```

```python
sales_occupation=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x='Occupation',y='Amount',data=sales_occupation)
```

From the BAove graph we can see that most of the buyers are from It sector, Healthcare sector and Aviation sector

```python

sns.set(rc={'figure.figsize':(24,5)})
ax =sns.countplot(data=df,x='Product_Category')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_food=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x='Product_Category',y='Amount',data=sales_food)
```

 From the above graphs we can see that most of sold products are from food, Footwear and Electronics category 

```python
sales_product=df.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x='Product_ID',y='Orders',data=sales_product)
```

```python
sales_product=df.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x='Product_ID',y='Orders',data=sales_product)
```
